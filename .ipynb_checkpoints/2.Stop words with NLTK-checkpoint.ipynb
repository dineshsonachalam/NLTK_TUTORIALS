```python
"""
    set() -> (i)   Unordered collection of items
             (ii)  Every element is unique(no duplicate)
             (iii) Set immutable(cannot be changed)
             (iv)  Permorms mathematical operation such as (union,intersection,symmetric difference)
"""
# Creation of a set
my_set_1 = {1,2,3}
# set of mixed datatypes
my_set_2 = {1.0,"Hello,(1,2,3)"}
print("Set1:",my_set_1)
print("Set2:",my_set_2,"\n")


# Set do not have any duplicate elements
my_set3 = {1,2,3,4,3,2}
print(my_set3)
my_set4 = set([1,2,3,2])
print(my_set4,"\n")

# Difference between dictionary and set()
a = {}
# Checking data type of a
print(type(a))
a = set()
# Checking data type of a
print(type(a))
```

```python
# This is an example Filtration
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

example_sent = "This is a sample sentence, showing off the stop words filtration."

word_tokens = word_tokenize(example_sent)
print("1. Tokenizing(Splitting) the string in the form of words and storing in a list:\n")
print(word_tokens,"\n\n\n")

stop_words = stopwords.words('english') # Now stop words will be a default list[] 
print(stop_words)
print("\n\n\n")
stop_words = set(stopwords.words('english')) # Now stop words will be set() ->words are immutable and no duplicates present
print(stop_words)



# Now creation of empty list
words_filtered = []
# Now removing the Stop words(Useless Words) 

for w in word_tokens:
    if(w not in stop_words):
        words_filtered.append(w)
        
print("\nOrginal Message:")
print(example_sent,)
print("\nFiltered Message:")
print(words_filtered)
```
